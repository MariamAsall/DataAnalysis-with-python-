# Data Preprocessing


```python
# new column >> sales = Price*Quantity 

All['Sales'] = All['Price']*All['Quantity']
```

```python
All.columns
```

```python
# Region >> UpperCase 
All.Region = All['Region'].str.upper()
All.Region
```

```python
# strip() >> like trim 
All.Product = All.Product.str.strip()
All.Product
```

```python
# missingg values 
All.isna().sum()
```

```python
# check duplicates
All.duplicated().sum()
```

```python
All.Date.duplicated().sum()
```

```python
# remove this 1 duplicate 
All = All.drop_duplicates()
All
```

# Analyze Data 

```python
# summtion of all sales 
All['Sales'].sum() 
```

```python
# partion this number for each region
All.groupby('Region')['Sales'].sum()  
```

```python
# Total Sales by Region 
All.groupby(['Region'])['Sales'].sum() 
```

```python
# Avg. Quantity  by Quantity 
All.groupby('Quantity')['Quantity'].mean()
```

```python
# Orders by Segment 
# groupby >> اجمع ال colتحت بعض و اشوف عددهم 
All.groupby('Segment')['Product'].count() 
```
